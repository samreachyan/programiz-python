```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Example of function y
### $$f(x) = x^2 + x + 1$$


```python
def f(x):
    return x**2 + x + 1
```

```python
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
plt.figure(figsize=[10, 5])
plt.title('Table function F(x) = x^2 + x + 1', fontsize=15)
plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.plot(x_1, f(x_1))
plt.show()
```

## Derivate or Plot of function

```python
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)

plt.title('Table function F(x) = x^2 + x + 1', fontsize=15)
plt.xlabel('x', fontsize=14)
plt.ylabel('df(x)', fontsize=14)
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.plot(x_1, f(x_1), color='green', linewidth=3)

plt.subplot(1, 2, 2)
plt.title('Table function df(x) = x^2 + x + 1', fontsize=15)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)
plt.plot(x_1, df(x_1), color='green', linewidth=3)

plt.show()
```

```python
# Gradient descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.0001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient

    x_list.append(new_x)
    slope_list.append(df(new_x))
    step_size = abs(new_x-previous_x)
    # print('Step size : ', step_size)
    if step_size < precision:
        print('Time running in precision: ', n)
        break

print('Local minize occure at : ', new_x)
print('Slop at this point : ', df(new_x))
print('Cost function : ', f(new_x))
```

```python
# Super cost function by points
plt.figure(figsize=[15, 5])

# chart 1
plt.subplot(1, 2, 1)
plt.title('Table function F(x) = x^2 + x + 1', fontsize=15)
plt.xlabel('x', fontsize=14)
plt.ylabel('df(x)', fontsize=14)
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.plot(x_1, f(x_1), color='green', linewidth=3)
value = np.array(x_list)
plt.scatter(x_list, f(value), color='blue', linewidths=1, s=100, alpha=0.3)

# chart 2
plt.subplot(1, 2, 2)
plt.title('Table function df(x)', fontsize=15)
plt.grid()
plt.xlim(-1, 3)
plt.ylim(-1, 6)
plt.plot(x_1, df(x_1), color='green', linewidth=3)
plt.scatter(x_list, slope_list, color='blue', s=100, alpha=0.3)

plt.show()
```

## $$ g(x) = x^4 - 4x^2 + 5 $$

```python
# Make some data
x_2 = np.linspace(-2, 2, 1000)

# fuction
def g(x):
    return x**4 - 4*x**2 + 5

# derivate
def dg(x):
    return 4*x**3 - 8*x
```

```python
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)

plt.title('Table function G(x)', fontsize=15)
plt.xlabel('x', fontsize=14)
plt.ylabel('dg(x)', fontsize=14)
plt.grid()
plt.xlim(-2, 2)
plt.ylim(0.5, 5.5)
plt.plot(x_2, g(x_2), color='green', linewidth=3)

plt.subplot(1, 2, 2)
plt.title('Table function dg(x)', fontsize=15)
plt.grid()
plt.xlim(-2, 2)
plt.ylim(-6, 8)
plt.plot(x_2, dg(x_2), color='green', linewidth=3)

plt.show()
```
