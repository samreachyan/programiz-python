```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Example of function y
### $$f(x) = x^2 + x + 1$$


```python
def f(x):
    return x**2 + x + 1
```

```python
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
plt.figure(figsize=[10, 5])
plt.title('Table function F(x) = x^2 + x + 1', fontsize=15)
plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.plot(x_1, f(x_1))
plt.show()
```

## Derivate or Plot of function

```python
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)

plt.title('Table function F(x) = x^2 + x + 1', fontsize=15)
plt.xlabel('x', fontsize=14)
plt.ylabel('df(x)', fontsize=14)
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.plot(x_1, f(x_1), color='green', linewidth=3)

plt.subplot(1, 2, 2)
plt.title('Table function df(x) = x^2 + x + 1', fontsize=15)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)
plt.plot(x_1, df(x_1), color='green', linewidth=3)

plt.show()
```
